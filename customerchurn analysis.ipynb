```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:/ffff/Customer Churn.csv")
df.head()
```

```python
df.info()
```

```python
# replacing blanks with 0
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
# converting datatype
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
# converting 0 and 1 value of Senior Citizen to yes or no.
def conv(value):
    if value ==1:
        return "yes"
    else:
        return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df
```

```python
plt.figure(figsize = (3,4))
a = sns.countplot(x = 'Churn', data = df)

a.bar_label(a.containers[0])
plt.title("Count of Customers by Churn", fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby('Churn').agg({'Churn': 'count'})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churn Customer", fontsize = 10)
plt.show()
```

```python
# from the given pie chart we can conclude that 26.54% of our customers have churned.
# now let's explore the reason behind it. 
```

```python
plt.figure(figsize = (3,4))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure", data = df, bins= 50, hue = 'Churn')
plt.show()
```

```python

# people who have used our services for long time have stayed and people have used our services for 1 or 2 months have churn
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Contract", data = df, hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
  

```

```python
# people who have month to month contract are likely to churn then from those who have 1 or 2 years of contact.
```

```python
# Define the columns for count plots
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up subplots (3 rows, 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(18, 12))  # Adjust grid based on 9 plots

# Flatten axes for easy iteration
axes = axes.flatten()

# Create count plots for each column
for i, col in enumerate(columns):
    sns.countplot(x=df[col], hue=df['Churn'], ax=axes[i])  # Add 'Churn' hue
    axes[i].set_title(f"Count of {col} by Churn")  # Add title
    axes[i].bar_label(axes[i].containers[0])  # Add labels on bars
    if len(axes[i].containers) > 1:  # Ensure labels for both Churn categories
        axes[i].bar_label(axes[i].containers[1])

# Adjust layout
plt.tight_layout()
plt.show()

```

```python
# Customers with no additional services (like online security, tech support, or streaming) show higher churn rates. In contrast, those subscribed to multiple services tend to stay.
```
